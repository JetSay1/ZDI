```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/sampledatafoodsales/sampledatafoodsales.csv')
df.drop('Unnamed: 7', axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(df['Product'], df['Quantity'])
ax.set(xlabel="Products",
       ylabel="Quantity",
       title="Максимальне замовлення для кожного продукту")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(df.groupby('City')['Quantity'].mean().index, df.groupby('City')['Quantity'].mean())
ax.set(xlabel="City",
       ylabel="Quantity",
       title="Середня кількість замовлень товарів на місто")
plt.show()
```

```python
region_sum = df.groupby('Region')['Quantity'].sum()

fig, ax = plt.subplots(figsize=(7, 8))
ax = region_sum.plot.bar()
ax.set_title('Кількість замовлень на регіон')
plt.show()
```

```python
pd.plotting.scatter_matrix(df, figsize=(10, 8))
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))
ax[0].boxplot(df['Quantity'])
ax[1].boxplot(df['UnitPrice'])
ax[0].set_title('Quantity')
ax[1].set_title('UnitPrice')
fig.suptitle("Quantity and UnitPrice boxplots")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))
ax[0].violinplot(df['Quantity'])
ax[1].violinplot(df['UnitPrice'])
ax[0].set_title('Quantity')
ax[1].set_title('UnitPrice')
plt.show()
```

```python
city_count = df['City'].value_counts()
product_count = df['Product'].value_counts()
region_count = df['Region'].value_counts()

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 6))
# ax.violinplot([city_count, product_count, df['Quantity']])
ax[0].violinplot(city_count, showmedians=True)
ax[0].set_title('City count')
ax[1].violinplot(product_count, showmedians=True)
ax[1].set_title('Product count')
ax[2].violinplot(df['Quantity'], showmedians=True)
ax[2].set_title('Quantity')
fig.suptitle("Violin Plots from City Product and Quantity columns")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.violinplot(x='Quantity', y='Product', data=df, palette='Set3')
ax.set_title("Скрипкова діаграма кількості для кожного продукту")
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.swarmplot(x="Quantity", y="Category", data=df, hue='Region').set_title("Кількість замовлень для кожної категорії")
plt.show()
```
